```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "duration_ms": [210000, 180000, 240000, 300000, 190000, 220000, 260000, 250000]
})

# Write code below to create and display a histogram showing the distribution of song durations
# Use 5 bins, 
# add a title "Distribution of Song Durations"
# label the axes "Duration" (x) and "Count" (y)

#YOUR CODE GOES HERE!! NOTE: If the axes and title labels do not match exactly, the code will not pass. Keep this simple do not do more than is required. 
plt.hist(df["duration_ms"], bins=5)
plt.title("Distribution of Song Durations")
plt.xlabel("Duration")
plt.ylabel("Count")
plt.show()
# REMEMBER TO EXECUTE THE CODE BEFORE MOVING ON TO THE NEXT QUESTION, THIS ALLOWS YOU TO CHECK IF YOU GOT IT RIGHT
```
