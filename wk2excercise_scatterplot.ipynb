```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "energy": [0.6, 0.8, 0.5, 0.9, 0.7, 0.65],
    "valence": [0.5, 0.7, 0.4, 0.8, 0.6, 0.55]
})

# Write code below to create and display a scatter plot of energy vs valence
# Label the axes "Energy" (x) and "Valence" (y)
# Title the chart "Energy vs Valence"

#YOUR CODE GOES HERE!! NOTE: If the axes and title labels do not match exactly, the code will not pass. Keep this simple do not do more than is required. 
plt.scatter(df["energy"], df["valence"])
plt.xlabel("Energy")
plt.ylabel("Valence")
plt.title("Energy vs Valence")
plt.show()
# REMEMBER TO EXECUTE THE CODE BEFORE MOVING ON TO THE NEXT QUESTION, THIS ALLOWS YOU TO CHECK IF YOU GOT IT RIGHT
```
