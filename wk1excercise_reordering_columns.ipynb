```python
import pandas as pd

data = {
"Name": ["Alice", "Bob", "Charlie"],
"Dept": ["HR", "Finance", "IT"],
"Sal": [60000, 70000, 80000],
}

df = pd.DataFrame(data)
```

```python
df = df.rename(columns={'Dept': 'Department', 'Sal': 'Salary'})
```

```python
df = df[['Salary', 'Name', 'Department']]
```
